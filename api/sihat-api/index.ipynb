## glob pre-processing

```python
list = sorted(glob.glob("test/BraTS20_training_020/*.nii"))
```

```python
processed_images = []
```

```python
for img in range(len(list)):
    temp_image = nib.load(list[img]).get_fdata()
    temp_image = scaler.fit_transform(temp_image.reshape(-1, temp_image.shape[-1])).reshape(temp_image.shape)

    processed_images.append(temp_image)

stacked_images = np.stack(processed_images, axis=3)
stacked_images = stacked_images[56:184, 56:184, 13:141]

np.save('test/images/image_test.npy', stacked_images)
```

## custom data generator

```python
import os
import numpy as np

def load_img(img_dir, img_list):
    images = []
    
    for i, image_name in enumerate(img_list):    
        if (image_name.split('.')[1] == 'npy'):
            
            image = np.load(img_dir+image_name)
                      
            images.append(image)
    images = np.array(images)
    
    return(images)

def imageLoader(img_dir, img_list, batch_size=1):

    L = len(img_list)

    #keras needs the generator infinite, so we will use while true  
    while True:

        batch_start = 0
        batch_end = batch_size

        while batch_start < L:
            limit = min(batch_end, L)
                       
            X = load_img(img_dir, img_list[batch_start:limit])

            yield (X) #a tuple with two numpy arrays with batch_size samples     

            batch_start += batch_size   
            batch_end += batch_size
```

```python
test_dir = "test/images/"
test_img_list = os.listdir(test_dir)
```

```python
test_img_datagen = imageLoader(test_dir, test_img_list)
```

```python
img = test_img_datagen.__next__()
```
